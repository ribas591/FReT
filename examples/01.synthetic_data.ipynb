# FReT on Synthetic Data

This notebook demonstrates the FReT (Forecasting using Recurrence Topology) algorithm on a synthetic time series — a sum of two sine waves. The algorithm learns from the training data and forecasts the next 100 time steps.

```python
import sys

sys.path.insert(0, "..")

import numpy as np
import matplotlib.pyplot as plt

from fret import FReT_forecast, mae
```

# Sample data

```python
t_step = np.arange(0.1, 70.1, 0.1)
data_tot = np.cos(2 * np.pi * t_step / 3) + 0.75 * np.sin(2 * np.pi * t_step / 5)
plt.plot(data_tot, linewidth=2)
plt.show()
```

```python
x_train = data_tot[:600].reshape(-1, 1)
x_test = data_tot[600:700].reshape(-1, 1)
```

# Forecast 

```python
forecast = FReT_forecast(x_train, forecast_horizon=100)
```

```python
forecast
```

# Plot data

```python
plt.figure(figsize=(12, 6))
plt.plot(x_test, linewidth=4, color="black", label="Test Data")
plt.plot(forecast, "--", linewidth=4, color="red", label="FReT")
plt.text(
    0.01,
    0.99,
    f"MAE = {np.round(mae(x_test, forecast), 2)}",
    fontsize=14,
    ha="left",
    va="top",
    transform=plt.gca().transAxes,
)
plt.legend(loc="lower left")
```
